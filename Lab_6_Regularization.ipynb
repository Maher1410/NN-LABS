```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/LABNN/telco_churn.csv')
print(df.head())
df.info()
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan)
df['TotalCharges'] = df['TotalCharges'].astype('float')
df.info()
```

```python
import numpy as np

df['Churn'] = np.where(df['Churn'] == 'Yes', 1, 0)

def convert_categories(cat_list):
    for col in cat_list:
        df[col] = df[col].astype('category')
        df[f'{col}_cat'] = df[f'{col}'].cat.codes

category_list = ['gender', 'Partner', 'Dependents', 'PhoneService',
                 'MultipleLines', 'InternetService',
                 'OnlineSecurity', 'OnlineBackup',
                 'DeviceProtection', 'TechSupport', 'StreamingTV',
                 'StreamingMovies', 'Contract', 'PaperlessBilling',
                 'PaymentMethod']
convert_categories(category_list)

print(df.head())
df['TotalCharges']
```

```python
X = df[['gender_cat', 'Partner_cat', 'Dependents_cat', 'PhoneService_cat', 'MultipleLines_cat',
        'OnlineSecurity_cat', 'OnlineBackup_cat', 'DeviceProtection_cat', 'TechSupport_cat',
        'StreamingMovies_cat', 'Contract_cat', 'PaperlessBilling_cat', 'PaymentMethod_cat',
        'TotalCharges', 'SeniorCitizen', 'tenure']]

y= df['Churn']
```

```python
seed = 7
np.random.seed(seed)

from sklearn.model_selection import train_test_split

X_train, X_test_hold_out, y_train, y_test_hold_out = train_test_split(X, y, test_size=0.2, random_state=seed)
X_train
```

```python
X_train = np.asarray(X_train).astype(np.float32)
y_train = np.asarray(y_train).astype(np.float32)

X_test_hold_out = np.asarray(X_test_hold_out).astype(np.float32)
y_test_hold_out = np.asarray(y_test_hold_out).astype(np.float32)
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.metrics import accuracy_score

model = Sequential()
model.add(Dense(X_train.shape[1], input_shape=(X_train.shape[1],), kernel_initializer='normal', activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='softmax'))
model.compile(optimizer = 'adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=20)
```

```python
y_pred = model.predict(X_test_hold_out)
y_pred = np.where(y_pred > 0.5, 1, 0)
print("Accuracy: ", accuracy_score(y_pred, y_test_hold_out))
```

```python
from tensorflow.keras import regularizers

model_lasso = Sequential()
model_lasso.add(Dense(X_train.shape[1], input_shape=(X_train.shape[1],), kernel_initializer='normal', activation='relu', kernel_regularizer=regularizers.l1(1e-6)))
model_lasso.add(Dense(32, activation='relu'))
model_lasso.add(Dense(32, activation='relu'))
model_lasso.add(Dense(1, activation='sigmoid'))
model_lasso.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_lasso.fit(X_train, y_train, epochs=20)
```

```python
y_pred = model_lasso.predict(X_test_hold_out)
y_pred = np.where(y_pred > 0.5, 1, 0)
print("Accuracy With Lasso: ", accuracy_score(y_pred, y_test_hold_out))
```

```python
model_ridge = Sequential()
model_ridge.add(Dense(X_train.shape[1], input_shape=(X_train.shape[1],), kernel_initializer='normal', activation='relu', kernel_regularizer=regularizers.l2(1e-6)))
model_ridge.add(Dense(32, activation='relu'))
model_ridge.add(Dense(32, activation='relu'))
model_ridge.add(Dense(1, activation='sigmoid'))
model_ridge.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_ridge.fit(X_train, y_train, epochs=20)
```

```python
y_pred = model_ridge.predict(X_test_hold_out)
y_pred = np.where(y_pred > 0.5, 1, 0)
print("Accuracy With Ridge: ", accuracy_score(y_pred, y_test_hold_out))
```
