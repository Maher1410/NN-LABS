```python
import numpy
from keras.models import Sequential
from keras.layers import Dense
from keras.datasets import mnist
import matplotlib.pyplot as plt
from keras.utils import to_categorical
```

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
```

```python
#load data
(X_train, y_train), (X_test,y_test) = mnist.load_data()
```

```python
#plot 4 images as gray scale
plt.subplot(221)
plt.imshow(X_train[0])
plt.subplot(222)
plt.imshow(X_train[1])
plt.subplot(223)
plt.imshow(X_train[2])
plt.subplot(224)
plt.imshow(X_train[3])
#show the plot
plt.show()
```

```python
#show number of training sample
X_train.shape
```

```python
num_pixels = X_train.shape[1] * X_train.shape[2]
X_train = X_train.reshape(X_train.shape[0], num_pixels)
X_test = X_test.reshape(X_test.shape[0], num_pixels)
```

```python
X_train.shape
```

```python
#Normalize inputs from 0-255 to 0-1
X_train = X_train / 255
X_test = X_test / 255
```

```python
y_test
```

```python
y_train = to_categorical (y_train)
y_test = to_categorical (y_test)
```

```python
num_classes = y_test.shape[1]
```

```python
num_classes
```

```python
#define basline model
def baseline_model():
# create model
  model = Sequential ()
  model.add(Dense(num_pixels, input_dim=num_pixels , activation= 'relu'))
  model.add(Dense(200,activation='relu'))
  model.add(Dense(num_classes , activation = 'softmax'))
# Compile model
  model.compile(loss='categorical_crossentropy' , optimizer='adam' , metrics=['accuracy'])
  return model
```

```python
#build the model
model = baseline_model()
```

```python
#Train the model
model.fit (X_train, y_train, epochs=10, batch_size=200)
```

```python
#evaluate the model with testing data
test_loss, test_acc = model.evaluate(X_test,y_test)
```

```python
print ('test_acc:', test_acc)
```

```python
y_predicted = model.predict(X_test)
```

```python
import numpy as np
y_predicted=[np.argmax(x) for x in y_predicted]
```

```python
y_predicted
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns

# Calculate confusion matrix
conf_matrix = confusion_matrix(y_test.argmax(axis=1), y_predicted)

# Calculate accuracy
accuracy = accuracy_score(y_test.argmax(axis=1), y_predicted)

print("Accuracy:", accuracy)

# Plot confusion matrix
plt.figure(figsize=(10,8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=range(10), yticklabels=range(10))
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Find the predicted label for the first image
predicted_value = y_predicted[0]
desired_value = y_test[0]

print("Predicted value for the first image:", predicted_value)
print("Desired value for the first image:", desired_value)
```
